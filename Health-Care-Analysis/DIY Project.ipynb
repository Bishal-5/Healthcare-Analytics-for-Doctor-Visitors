# Information form the data set

Plot for Visits vs Income

```python
sns.barplot(x=df['visits'],y=df['income'],data=df)
```

Find out the total number of people based on their count of illness

```python
df["illness"].value_counts()
```

Analyze the maximum, minimum and medium income

```python
x = list(df.income)
plt.boxplot(x)
plt.title('Maximum, Minimum & Medium Income')
plt.show()
```

Find out the no. of days of reduced activity of male & female separately due to illness

```python
df.groupby(['gender','reduced']).mean()
```

Visualize is there any missing values in the dataset based on a heat map

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

Analyze how the income of a patient affects the no. of visits to the hospital.

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

Count the number of male & female affected by illness

```python
sns.histplot(df.gender,bins=2)
```

Plot a horizontal bar chart to analyze the reduced days of activity illness based on gender

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#Creating the bar chart
plt.barh(db['gender'],db['reduced'],color = ['cornflowerblue','lightseagreen'])
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```

Visualize the percentage of people getting govt health Insurance due to low income, due to old age & also the percentage of people having private health insurance

```python
# % of people getting govt Insurance due to low income
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt Insurance")
plt.show()
# % of people having private insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people having private insurance")
plt.show()
# % of people getting govt insurance due to old age, disability or veteran status
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt insurance due to old age, disability or veteran status")
plt.show()
```
